```python
# Simple Calculator
# Write a program to perform addition, subtraction, multiplication, or division based on user choice.

num_1 = int(input("Enter the first number:"))
num_2 = int(input("Enter the second number: "))

while True:
  op = int(input("Enter the operation number to perform:\n"
      "1.Addition\n"
      "2.Subtraction\n"
      "3.Multiplication\n"
      "4.Divide\n"))
  if op == 1:
    Add = num_1 + num_2
    print(f"{num_1} + {num_2} = {Add}")
  elif op == 2:
     Sub = num_1 - num_2
     print(f"{num_1} - {num_2} = {Sub}")
  elif op == 3:
     Mul = num_1 * num_2
     print(f"{num_1} * {num_2} = {Mul}")
  elif op == 4:
     if num_2 != 0:
      Div = num_1 / num_2
      print(f"{num_1} / {num_2} = {Div}")
     else:
      print("Cannot divide by zero")
  else:
     print("Invalid operation")
     break
```

```python
# Electricity Bill
# Write a program to calculate electricity bill:

units = int(input("Enter your electricity units: "))

if units <= 100:
    bill = units * 1
elif units <= 200:
    bill = units * 2
else:
    bill = units * 3

print(f"Your electricity bill is {bill}")

```

```python
# Sum of Numbers
# Write a program to find the sum of first n natural numbers.

num = int(input("Enter the number to count: "))

total = 0

for i in range(1, num+1):
  total += i
print(f"Total of first {num} is {total}")
```
